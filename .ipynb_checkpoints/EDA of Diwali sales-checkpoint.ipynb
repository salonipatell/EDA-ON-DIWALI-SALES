# Exploratory Data Analysis

# Analyse Number of Customers by Gender

```python
# sns.histplot(data['Gender'])
sns.set(rc={'figure.figsize':(7,5)})
ax=sns.countplot(x='Gender', data=data)
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Number of Customers by Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

From the abbove graph we can see that Number of females customers are more than males

# Analyse purchasing power by gender

```python
data.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by="Amount",ascending=False)
```

```python
sales_gen=data.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by="Amount",ascending=False)
sales_bar=sns.barplot(x='Gender',y='Amount',data=sales_gen)
plt.title("Purchased By Gender")
```

From above graph we can see that most of the buyers are female and even purchasing power of female are greater than male

```python
# sort the age group in asc order
data.sort_values('Age Group', ascending=True,inplace=True)    
```

```python
data.head()
```

# Analyse Number of Customers by Age Group

```python
age_g=sns.countplot(data=data,x='Age Group',hue='Gender')
for i in age_g.containers:
    age_g.bar_label(i)
plt.title(' Number of Customers by Age Group')
```

# Amount Spent ByAgeGroup

```python
group_amt=data.groupby(['Age Group'],as_index=False)['Amount'].sum()
sns.barplot(x='Age Group',y='Amount',data=group_amt)
```

From the above graph we can see that most of the buyers are of age group between 26-35 yrs Female

# Analyse orders by top 10 state

```python
 
orders_state=data.groupby('State',as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(24,6)})
sns.barplot(x='State',y='Orders',data=orders_state)
plt.title('TOP 10 STATES')
```

```python
#amount spent by top 10 state 
orders_state=data.groupby('State',as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(24,6)})
sns.barplot(x='State',y='Amount',data=orders_state)
plt.title(' Amount spent by top 10 states')
```

from above graphs we can see that most orders and high purchasing power are from uttarpradesh,maharashtra,karnatake respectively

# Analyse Number of customers by Martial Status

```python
sns.set(rc={'figure.figsize':(7,5)})
a=sns.countplot(data=data,x='Marital_Status')
for i in a.containers:
    a.bar_label(i)
    
```

```python
sales_m=data.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(7,5)})
sns.barplot(data=sales_m,x='Marital_Status',y='Amount',hue='Gender')
```

Most of the buyers are married women and they have purchasing power

# Analyse Number of cutomers by Occupation

```python
sns.set(rc={'figure.figsize':(26,6)})
k=sns.countplot(x='Occupation',data=data)
for i in k.containers:
    k.bar_label(i)
```

from above graphs we can see that most of the buyers from it sector ,healthcare and aviation

# Analyse top product  category

```python

pro=data.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(24,6)})
sns.barplot(x='Product_Category',y='Amount',data=pro)
```

```python
sns.set(rc={"figure.figsize":(28,6)})
a=sns.countplot(data=data,x='Product_Category')
for i in a.containers:
    a.bar_label(i)
```

most orders of clothing and apparel but  most amount generated from the food

```python
# Product id
pro=data.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(24,6)})
sns.barplot(x='Product_ID',y='Orders',data=pro)
```
